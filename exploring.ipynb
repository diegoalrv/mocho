```python
import geopandas as gpd
import pandas as pd
import os
```

```python
data_folder = "data/"
def load_data(data_folder):
    # scan all pkl files in the data folder
    files = [f for f in os.listdir(data_folder) if f.endswith('.pkl')]
    data = {}
    for f in files:
        df = pd.read_pickle(os.path.join(data_folder, f))
        data[f] = df
    return data

data = load_data(data_folder)
```
